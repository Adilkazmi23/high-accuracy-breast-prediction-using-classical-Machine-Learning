```python
df = df.drop(['id','Unnamed: 32'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
Labelencoder= LabelEncoder()
df['diagnosis']= Labelencoder.fit_transform(df['diagnosis'])
```

```python
plt.figure (figsize=(12,10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title('feature correlation heatmap')
plt.show()
```

```python
x = df.drop('diagnosis',axis=1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
log = LogisticRegression(random_state=0)
log.fit(x_train, y_train)
```

```python
tree =DecisionTreeClassifier(random_state = 0 , criterion ='entropy')
tree.fit(x_train, y_train)
```

```python
forest = RandomForestClassifier(random_state = 0 , criterion ='entropy', n_estimators=10)
forest.fit(x_train, y_train)
```

```python
log_train_prediction = log.predict(x_train)
tree_train_prediction = tree.predict(x_train)
forest_train_prediction = forest.predict(x_train)
print('Logistic Regression training accuracy', accuracy_score(log_train_prediction, y_train))
print('Decision Tree training accuracy', accuracy_score(tree_train_prediction, y_train))
print('Random Forest training accuracy', accuracy_score(forest_train_prediction, y_train))
```

```python
models = [log , tree , forest]
for i , model in enumerate(models):
    y_pred = model.predict(x_test)
    print (f'/nModel {i} / ( Accuracy: {accuracy_score(y_test , y_pred):.2f}')
    print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True , fmt='d' , cmap='Blues')
plt.title(f'confused matrix - model {i}')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import pickle
with open ('forest_cancer_model.pkl', 'wb') as file:
  pickle.dump(forest, file)
    
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler,file)
```
